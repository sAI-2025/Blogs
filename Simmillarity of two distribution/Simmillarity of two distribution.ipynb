### Computing Percentiles

```python
import numpy as no
def compute_percentiles(data):
    """
    Computes percentiles (1 to 100) for a given dataset using numpy.
    Rounds the percentile values to 1 decimal place.
    
    Parameters:
    data (list): The input dataset (unsorted or sorted).
    
    Returns:
    list: Percentile values rounded to 1 decimal place.
    """
    # Ensure the data is a numpy array and sort it
    data = np.sort(data)
    # Compute percentiles 1 to 100 and round to 1 decimal place
    percentiles = [round(np.percentile(data, p), 1) for p in range(1, 101)]
    return percentiles
```

### QQ plots on X test and X val

```python
import matplotlib.pyplot as plt
def QQPlots(x_values, y_values, title="QQ Plot on Xtest and Xval"):
    # Check input lengths
    if len(x_values) != len(y_values):
        raise ValueError("Lengths of x_values and y_values must be the same")
    # Convert lists to numpy arrays
    x = np.array(x_values)
    y = np.array(y_values)

    # Scatter plot with unique colors
    colors = np.random.rand(len(x))  # Random colors
    plt.scatter(x, y, c=colors, cmap='viridis', label='Data Points', s=100)

    # Plot the y=x line
    x_line = np.linspace(min(x.min(), y.min()), max(x.max(), y.max()), 100)
    y_line = x_line
    plt.plot(x_line, y_line, color='blue', label='y=x', linestyle='--')

    # Add graph details
    plt.xlabel('Xtest')
    plt.ylabel('Xval')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
```
